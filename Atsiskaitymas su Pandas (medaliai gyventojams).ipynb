```python
import mysql.connector as CNT
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.cbook import get_sample_data
%matplotlib tk

M= pd.read_csv('Medals.csv')
C= pd.read_csv('Countries.csv')

MM=M[['Country_Name', 'Bendras']]
M2=MM.groupby(['Country_Name']).sum('Bendras').sort_values(by='Bendras', ascending=False).reset_index()
CC=C[['Country', 'Population']]
EF= pd.merge(M2, CC, how='inner', left_on='Country_Name' ,right_on= 'Country')

EF['Sk']= 1000000*EF['Bendras']/EF['Population']

Gal=EF[['Country', 'Bendras', 'Sk']].sort_values(by='Sk', ascending=False)
Sk=Gal['Sk'].values
Salis=Gal['Country'].values

x=[]
y=[]
c=[]   
a=0
for i in range(0,len(Salis)):
    a=a+1
    if i <50 or Salis[i] =='Lithuania' or Salis[i] =='Latvia' or Salis[i] =='Estonia':
        x.append(str(a)+' '+(Salis[i]))
        y.append(Sk[i])

for i in range(0,len(x)):     
    if  'Lithuania' in x[i]:
        c.append('#006A44')
    elif 'Latvia' in x[i]:
        c.append('#9D2235')
    elif 'Estonia' in x[i]:
        c.append('#0072ce')
    else:
        c.append('grey')

fig, g= plt.subplots()
a=g.bar(x, y, color=c)



g.bar_label(a, rotation=90, fmt='%.1f', label_type='center')
fig.suptitle('Medalių skaičius/ 1 milijonui gyventojų *', fontsize=14)


# g.set_xlabel('Šalies pavadinimas', fontsize=12) 
# g.set_ylabel('Medalių skaičius milijonui gyventojų', fontsize=12) 
g.set_xticks(x)
g.set_xticklabels(x, rotation=90)
plt.tight_layout()

fig.canvas.manager.set_window_title('Fig.3')
im = plt.imread(get_sample_data('C:\\Users\\artur\\OneDrive\\Paveikslėliai\\O4.png'))
newax = fig.add_axes([0, 0.9, 0.1, 0.1], zorder=0)
newax.imshow(im)
newax.axis('off')


tekstas='* Bendras visų iškovotų medalių (1992m-2020m vasaros olimpiadose) skaičius, tenkantis vienam milijonui gyventojų.'
plt.text(0.01, 0.1, tekstas, fontsize=10, style= 'italic', transform=plt.gcf().transFigure)
plt.show()

```
