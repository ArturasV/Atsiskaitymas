```python
import mysql.connector as CNT
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.cbook import get_sample_data
%matplotlib tk

DBworks= CNT.connect(
host='localhost',
user='root',
password='***',
port=3317,
database='works'
)
c=DBworks.cursor()

c.execute("select Country_name, sum(Bronze), Sum(Silver), sum(Gold), sum(Bendras) as sum from Medals where Year between 1896 and 1990 group by Country_name order by sum desc limit 10")
ats=c.fetchall()
c.execute("select Country_name, sum(Bronze), Sum(Silver), sum(Gold), sum(Bendras) as sum from Medals where Year between 1990 and 2022 group by Country_name order by sum desc limit 10")
ats1=c.fetchall()

Salis=[]
Medaliai=[]
ASB=[]
ASB2=[]
Colors= ['#cd7f32','#C0C0C0','#ffd700']
C=Colors*10

for i in range(0, len(ats)):
    Salis.append(ats[i][0])
    Medaliai.append(ats[i][4])
    ASB.append(ats[i][1:4])
for i in range(0,len(ASB)):
    ASB2.extend(ASB[i])

Salis1=[]
Medaliai1=[]
ASB1=[]
ASB3=[]
for i in range(0, len(ats1)):
    Salis1.append(ats1[i][0])
    Medaliai1.append(ats1[i][4])
    ASB1.append(ats1[i][1:4])
for i in range(0,len(ASB)):
    ASB3.extend(ASB[i])


fig, ax = plt.subplots(1,2)
ax[0].pie(Medaliai, labels=Salis, autopct='%1.1f%%', shadow=True, pctdistance=0.85, startangle=0)
ax[0].pie(ASB2, colors=C, radius=0.65, startangle=0)

ax[1].pie(Medaliai1, labels=Salis1, autopct='%1.1f%%', shadow=True, pctdistance=0.85, startangle=0)
ax[1].pie(ASB3, colors=C, radius=0.65,startangle=0)

centre_circle = plt.Circle((0,0),0.5, fc='white')
ax[0].add_artist(centre_circle)
centre_circle1 = plt.Circle((0,0),0.5, fc='white')

ax[1].add_artist(centre_circle1)

Bronza = mpatches.Patch(color='#cd7f32', label='Bronza')
Sidabras = mpatches.Patch(color='#C0C0C0', label='Sidabras')
Auksas= mpatches.Patch(color='#ffd700', label='Auksas')
ax[0].legend(handles=[Auksas,Sidabras,Bronza],loc='lower left')
ax[1].legend(handles=[Auksas,Sidabras,Bronza],loc='lower right')
ax[0].set_title("1896-1988 metais", fontweight='bold')
ax[1].set_title("1992-2020 metais", fontweight='bold')
ax[0].axis('equal')
ax[1].axis('equal')
fig.suptitle('Šalys, gavusios daugiausiai olimpinių medalių', fontsize=16)
plt.tight_layout()
fig.canvas.manager.set_window_title('Fig.2')

im = plt.imread(get_sample_data('C:\\ProgramData\\MySQL\\MySQL Server 8.0\\Uploads\\O4.png'))
newax = fig.add_axes([0, 0.9, 0.1, 0.1], zorder=0)
newax.imshow(im)
newax.axis('off')

plt.show()
```
